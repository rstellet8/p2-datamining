```python
sales = pd.read_csv("dados/sales data-set.csv")
sales.head()
```

```python
sales.info()
```

```python
sales["Date"] = pd.to_datetime(sales["Date"])
sales.info()
```

```python
mask = []
for i in sales["Date"]:
    if i.month == 12:
        mask.append(True)
    else:
        mask.append(False)

SalesAveWithoutDec = mean(sales.loc[~pd.Series(mask)]['Weekly_Sales'])
SalesAveDec = mean(sales.loc[mask]['Weekly_Sales'])

print(f"Média das vendas exceto dezembro: {SalesAveWithoutDec:.2f}")
print(f"Média das vendas no mês de dezembro: {SalesAveDec:.2f}")
```

```python
def mean(series):
    return sum(series)/len(series)

weeklySalesHolidayTrue = mean(sales.query("IsHoliday == True")["Weekly_Sales"])
print(weeklySalesHolidayTrue)
weeklySalesHolidayFalse = mean(sales.query("IsHoliday == False")["Weekly_Sales"])
print(weeklySalesHolidayFalse)
print((weeklySalesHolidayTrue - weeklySalesHolidayFalse) / weeklySalesHolidayFalse)
```

```python
plt.bar(x=["False", "True"], height=[weeklySalesHolidayFalse, weeklySalesHolidayTrue])
plt.title("Mean Weekly Sales")
plt.xlabel("Is Holiday?")
plt.show()
```

```python
for dept in sales["Dept"].unique():
    salesByDept = sales.query(f"Store == 1 and Dept == {dept}")
    salesByDept = salesByDept.sort_values("Date")
    plt.plot(salesByDept["Date"], salesByDept["Weekly_Sales"], label=dept)

plt.xticks([datetime.datetime(year, 1, 1) for year in range(2010, 2014)])

#plt.legend(loc="upper right", ncol=2)
#plt.ylim((0, 100000))
plt.show()
```
